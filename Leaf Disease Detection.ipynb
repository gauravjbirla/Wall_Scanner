# 5. Image Detection

```python
import os
```

```python
import glob
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
folder = "/content/drive/MyDrive/YOLO-LealDisease-Detection-main/testing"
# iterate over files in that directory
for images in glob.iglob(f'{folder}/*'):
    img = os.path.join('data', 'images', images)
    results = model(img)
    results.print()

    %matplotlib inline 
    plt.imshow(np.squeeze(results.render()))
    plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
